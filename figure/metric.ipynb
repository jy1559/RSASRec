```python
import matplotlib.pyplot as plt

# Create a new figure
fig, ax = plt.subplots(figsize=(10, 6))
ax.axis('off')  # Hide axes

# Define box style
box_style = dict(boxstyle="round,pad=0.5", fc="lightblue", ec="navy", lw=2)

# Add boxes with text (all in English)
session_box = ax.text(0.5, 0.9, "Session Items\n(Item metadata & interactions)",
                        ha="center", va="center", fontsize=12, bbox=box_style)

ideal_box = ax.text(0.25, 0.7, "Ideal Ranking\n(r_j = j)",
                    ha="center", va="center", fontsize=12, bbox=box_style)
pred_box = ax.text(0.75, 0.7, "Predicted Ranking\n(r_j: predicted order)",
                   ha="center", va="center", fontsize=12, bbox=box_style)

ideal_util_box = ax.text(0.25, 0.5, "Ideal Utility\nUideal = [1/(log(j+1))]",
                         ha="center", va="center", fontsize=12, bbox=box_style)
actual_util_box = ax.text(0.75, 0.5, "Actual Utility\nU = [w_j  1/(log(r_j+1))]",
                          ha="center", va="center", fontsize=12, bbox=box_style)

nsu_box = ax.text(0.5, 0.3, "NSU\n= U / Uideal", ha="center", va="center",
                  fontsize=12, bbox=box_style)

# Draw arrows connecting the boxes
# From Session Items to both Ideal and Predicted Ranking
ax.annotate("", xy=(0.25, 0.8), xytext=(0.5, 0.85),
            arrowprops=dict(arrowstyle="->", lw=2))
ax.annotate("", xy=(0.75, 0.8), xytext=(0.5, 0.85),
            arrowprops=dict(arrowstyle="->", lw=2))

# From Ideal Ranking to Ideal Utility
ax.annotate("", xy=(0.25, 0.62), xytext=(0.25, 0.68),
            arrowprops=dict(arrowstyle="->", lw=2))
# From Predicted Ranking to Actual Utility
ax.annotate("", xy=(0.75, 0.62), xytext=(0.75, 0.68),
            arrowprops=dict(arrowstyle="->", lw=2))

# Arrows from utilities merging into NSU box
ax.annotate("", xy=(0.35, 0.42), xytext=(0.25, 0.42),
            arrowprops=dict(arrowstyle="->", lw=2))
ax.annotate("", xy=(0.65, 0.42), xytext=(0.75, 0.42),
            arrowprops=dict(arrowstyle="->", lw=2))

plt.tight_layout()
plt.show()
```
